```python
from osgeo import gdal
#@title **height vs reflectance bands of the buffered region**{ vertical-output: true }


def extract_pixel_values(raster_path, latitudes, longitudes, band_index=1):
    dataset = gdal.Open(raster_path)
    band = dataset.GetRasterBand(band_index)
    transform = dataset.GetGeoTransform()

    pixel_values = []

    for lat, lon in zip(latitudes, longitudes):
        x_coord = int((lon - transform[0]) / transform[1])
        y_coord = int((lat - transform[3]) / transform[5])

        pixel_value = band.ReadAsArray(x_coord, y_coord, 1, 1)[0, 0]
        pixel_values.append(pixel_value)

    dataset = None
    return pixel_values

# Example usage
latitudes = lat_1_buf
longitudes = lon_1_buf


raster_path = "/content/MOD09A1.061_sur_refl_b01_doy2023001_aid0001.tif" #@param {type:"string"}
pixel_values_b1 = extract_pixel_values(raster_path, latitudes, longitudes, band_index=1)
raster_path = "/content/MOD09A1.061_sur_refl_b02_doy2023001_aid0001.tif" #@param {type:"string"}
pixel_values_b2 = extract_pixel_values(raster_path, latitudes, longitudes, band_index=1)
raster_path = "/content/MOD09A1.061_sur_refl_b03_doy2023001_aid0001.tif" #@param {type:"string"}
pixel_values_b3 = extract_pixel_values(raster_path, latitudes, longitudes, band_index=1)
raster_path = "/content/MOD09A1.061_sur_refl_b04_doy2023001_aid0001.tif" #@param {type:"string"}
pixel_values_b4 = extract_pixel_values(raster_path, latitudes, longitudes, band_index=1)
raster_path = "/content/MOD09A1.061_sur_refl_b05_doy2023001_aid0001.tif" #@param {type:"string"}
pixel_values_b5 = extract_pixel_values(raster_path, latitudes, longitudes, band_index=1)
raster_path = "/content/MOD09A1.061_sur_refl_b06_doy2023001_aid0001.tif" #@param {type:"string"}
pixel_values_b6 = extract_pixel_values(raster_path, latitudes, longitudes, band_index=1)
raster_path = "/content/MOD09A1.061_sur_refl_b07_doy2023001_aid0001.tif" #@param {type:"string"}
pixel_values_b7 = extract_pixel_values(raster_path, latitudes, longitudes, band_index=1)
# print("Pixel values:", pixel_values[-10:])

# Combine the arrays into a single NumPy array
point_cloud_array = np.column_stack((latitudes, longitudes, height_1_buf, pixel_values_b1, pixel_values_b2, pixel_values_b3, pixel_values_b4, pixel_values_b5, pixel_values_b6, pixel_values_b7))

# Save the NumPy array to a file using np.save()
np.save('point_cloud.npy', point_cloud_array)

import numpy as np
import matplotlib.pyplot as plt

# Your data array
data = point_cloud_array

# Extract third and fourth columns

y = data[:, 2]          #height
b1 = data[:, 3]/10000    #reflectance
b2 = data[:, 4]/10000    #reflectance
b3 = data[:, 5]/10000    #reflectance
b4 = data[:, 6]/10000    #reflectance
b5 = data[:, 7]/10000    #reflectance
b6 = data[:, 8]/10000    #reflectance
b7 = data[:, 9]/10000    #reflectance

# Create a scatter plot
plt.scatter(b1, y, s=1, c='blue', alpha=0.3, label='band b1')  # Adjust s, c, and alpha as needed
plt.scatter(b2, y, s=1, c='red', alpha=0.3, label='band b2')  # Adjust s, c, and alpha as needed
plt.scatter(b3, y, s=1, c='yellow', alpha=0.3, label='band b3')  # Adjust s, c, and alpha as needed
plt.scatter(b4, y, s=1, c='cyan', alpha=0.3, label='band b4')  # Adjust s, c, and alpha as needed
plt.scatter(b5, y, s=1, c='pink', alpha=0.3, label='band b5')  # Adjust s, c, and alpha as needed
plt.scatter(b6, y, s=1, c='green', alpha=0.3, label='band b6')  # Adjust s, c, and alpha as needed
plt.scatter(b7, y, s=1, c='black', alpha=0.3, label='band b7')  # Adjust s, c, and alpha as needed
plt.xlabel('reflectance')
plt.ylabel('height')
plt.title('Scatter Plot of height vs reflectance')
plt.xlim(-0.01, 1.6)  # Set x-axis limits
plt.legend()  # Show legend

plt.grid(True)
plt.show()
```

```python
from osgeo import gdal

def extract_pixel_values(raster_path, latitudes, longitudes, band_index=1):
    dataset = gdal.Open(raster_path)
    band = dataset.GetRasterBand(band_index)
    transform = dataset.GetGeoTransform()

    pixel_values = []

    for lat, lon in zip(latitudes, longitudes):
        x_coord = int((lon - transform[0]) / transform[1])
        y_coord = int((lat - transform[3]) / transform[5])

        pixel_value = band.ReadAsArray(x_coord, y_coord, 1, 1)[0, 0]
        pixel_values.append(pixel_value)

    dataset = None
    return pixel_values

# Example usage
raster_path = "/content/MOD09A1.061_sur_refl_b04_doy2023001_aid0001.tif"
latitudes = lat_1_buf
longitudes = lon_1_buf

pixel_values = extract_pixel_values(raster_path, latitudes, longitudes, band_index=1)
print("Pixel values:", pixel_values[-10:])

# Combine the arrays into a single NumPy array
point_cloud_array = np.column_stack((latitudes, longitudes, height_1_buf, pixel_values))

# Save the NumPy array to a file using np.save()
np.save('point_cloud.npy', point_cloud_array)
```
